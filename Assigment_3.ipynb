```python
df = pnd.read_csv("grades.csv", sep=" ", skiprows=1, usecols=[1,2], names = ["math_SAT", "verb_SAT"])
```

```python
d = {"X": df.loc[0:104, "math_SAT"], 
     "Y": df.loc[0:104, "verb_SAT"],
     "X*Y": df.loc[0:104, "math_SAT"] * df.loc[0:104, "verb_SAT"],
     "X/Y": df.loc[0:104, "math_SAT"] / df.loc[0:104, "verb_SAT"],
     "X-Y": df.loc[0:104, "math_SAT"] - df.loc[0:104, "verb_SAT"]}

dfEnd = pnd.DataFrame(d, columns = ["X", "Y", "X*Y", "X/Y", "X-Y"])
dfEnd.corr(method = 'pearson')
```

<p><b>Задача 2.</b> Графики</p>

<p>Постройте графики в виде матрицы 2x2 для $(X,Y)$, $(X,X\cdot Y)$, $(X,X/Y)$ и $(X,X-Y)$. Добавьте в них прямые регрессии для каждого сочетания. Укажите каждого графика его название по сочетанию (X,Y и др.), а в легенде, какие данные он содержит.</p>

<p>Наклон и пересечение для прямых можно определить следующим образом:</p>

slope, intercept, r, p_value, std_err = stats.linregress(<X>, <Y>)<br>
print(slope, intercept)

```python
plt.figure(1, figsize=[17,17])

#----------------------------------------
x = dfEnd.X.values.astype(np.float32)
y = dfEnd.Y.values.astype(np.float32)
x_train = np.array([x[:]]).T
y_train = np.array([y[:]]).T
linRegModel_y = LinearRegression()
lrModelTr_y = linRegModel_y.fit(x_train, y_train)
f_x = lambda x : x * lrModelTr_y.coef_ + lrModelTr_y.intercept_

plt.subplot(2, 2, 1)
plt.title('X, Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(dfEnd.X, dfEnd.Y, "o")
blue_patch = mpatches.Patch(color='blue', label='X - high school GRA \nY - university GRA')
plt.legend(handles=[blue_patch], loc = 'lower right')
plt.plot(x_train, f_x(x_train), "-", color="red", lw=1)
plt.grid(True)

#----------------------------------------

x = dfEnd.X.values.astype(np.float32)
z = dfEnd.Y.values.astype(np.float32)
y = x*z
x_train = np.array([x[:]]).T
y_train = np.array([y[:]]).T
linRegModel_y = LinearRegression()
lrModelTr_y = linRegModel_y.fit(x_train, y_train)
f_x = lambda x : x * lrModelTr_y.coef_ + lrModelTr_y.intercept_

plt.subplot(2, 2, 2)
plt.title('X, X*Y')
plt.xlabel('X')
plt.ylabel('X*Y')
plt.plot(dfEnd.X, dfEnd.X*dfEnd.Y, "o")
blue_patch = mpatches.Patch(color='blue', label='X - high school GRA \nX * Y - product of high school and university GRA')
plt.legend(handles=[blue_patch], loc = 'lower right')
plt.plot(x_train, f_x(x_train), "-", color="red", lw=1)
plt.grid(True)


#----------------------------------------

x = dfEnd.X.values.astype(np.float32)
z = dfEnd.Y.values.astype(np.float32)
y = x/z
x_train = np.array([x[:]]).T
y_train = np.array([y[:]]).T
linRegModel_y = LinearRegression()
lrModelTr_y = linRegModel_y.fit(x_train, y_train)
f_x = lambda x : x * lrModelTr_y.coef_ + lrModelTr_y.intercept_

plt.subplot(2, 2, 3)
plt.title('X, X/Y')
plt.xlabel('X')
plt.ylabel('X/Y')
plt.plot(dfEnd.X, dfEnd.X/dfEnd.Y, "o")
blue_patch = mpatches.Patch(color='blue', label='X - high school GRA \nX / Y - quotient of high school and university GRA')
plt.legend(handles=[blue_patch], loc = 'lower right')
plt.plot(x_train, f_x(x_train), "-", color="red", lw=1)
plt.grid(True)


#----------------------------------------

x = dfEnd.X.values.astype(np.float32)
z = dfEnd.Y.values.astype(np.float32)
y = x-z
x_train = np.array([x[:]]).T
y_train = np.array([y[:]]).T
linRegModel_y = LinearRegression()
lrModelTr_y = linRegModel_y.fit(x_train, y_train)
f_x = lambda x : x * lrModelTr_y.coef_ + lrModelTr_y.intercept_

plt.subplot(2, 2, 4)
plt.title('X, X-Y')
plt.xlabel('X')
plt.ylabel('X-Y')
plt.plot(dfEnd.X, dfEnd.X-dfEnd.Y, "o")
blue_patch = mpatches.Patch(color='blue', label='X - high school GRA \nX - Y - difference between high school and university GPA')
plt.legend(handles=[blue_patch], loc = 'lower right')
plt.plot(x_train, f_x(x_train), "-", color="red", lw=1)
plt.grid(True)

#----------------------------------------
plt.show()
```
