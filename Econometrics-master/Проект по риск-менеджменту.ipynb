Приступим к прощету вариации дельта-нормальным методом:

Посчитаем 1-дневную дисперсию доходности портфеля:

```python
import scipy
Doli_m = Doli.as_matrix()
Covar_m = covar.as_matrix()
Mean_m = Mean.as_matrix()
H1 = []
H2 = []
H3 = []
H4 = []
H5 = []
for i in range(125):
    H1.append(np.dot(np.dot(Doli_m[i].T, Covar_m), Doli_m[i]))
    H2.append(np.sqrt(np.dot(np.dot(Doli_m[i].T, Covar_m), Doli_m[i])))
    H3.append(np.sqrt(10)*np.sqrt(np.dot(np.dot(Doli_m[i].T, Covar_m), Doli_m[i])))
    H4.append(scipy.stats.norm.ppf(0.99))
    H5.append(np.dot(Mean_m[0], Doli_m[i].T))
one = pd.DataFrame({'Var': H1, 'sigma': H2, 'ten_days_sigma': H3, 'Quintil': H4, 'E': H5}, columns
                       = ['Var', 'sigma', 'ten_days_sigma', 'Quintil', 'E'])
one.head()
```

```python
two = pd.DataFrame(columns= ['ten_days_VAR', 'ten_days_VAR_abs'])
two.ten_days_VAR = -(one.ten_days_sigma*one.Quintil + one.E)
two.ten_days_VAR_abs = -(one.ten_days_sigma*one.Quintil + one.E)*SUM_VL.SUM
two.head(10)
```

<h3 align=center>Backtesting of delta-normal

```python
thr = pd.DataFrame({'ten_days_VAR_abs': two.ten_days_VAR_abs, 'LOSE_WIN_10': LOSE.LOSE_WIN_10},
                   columns= ['ten_days_VAR_abs', 'LOSE_WIN_10', 'Indication'])
for i in range(116):
    if thr.ten_days_VAR_abs[i]>thr.LOSE_WIN_10[i]:
        thr.Indication[i] = 1
    else:
        thr.Indication[i] = 0
thr
```

```python
plt.plot(thr.ten_days_VAR_abs)
plt.plot(thr.LOSE_WIN_10)
plt.show()
```
